###ROAS(Return On Advertising Spend)
*   Definition: Marketing indicator that shows return on ad spend
*   The way to calculate:ROAS = Revenue from Ads / Ad Spend



*   Importance

    *   High ROAS: Refers to a high return on ad spend. Indicates the effectiveness of an ad campaign
    *   Low ROAS: Refers to a low return on ad investment. Suggests that your ad strategy needs improvement
