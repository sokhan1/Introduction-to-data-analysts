#Filtering


*   Filtering in pd.DataFrame is the process of leaving only rows that meet a condition




```python
import pandas as pd
data = {
    'Name':['A','B','C','D'],
    'Age':[25,30,35,40],
    'City':['New York','Los Angeles','New York', 'Chicago']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
con = df.Age>30
df['age_over_30'] = con
df
```

```python
df[con]
```

```python
df[df.Age>30]
```

#between(), isin(), isna()

```python
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
df.head() #20 years old or older 22 years old or younger # between(20,22)
```

```python
con = df.age.between(20,22) #method
```

```python
df[con].age.value_counts()
```

```python
df.embarked.value_counts()
```

```python
con2 = df.embarked.isin(['S','C'])
df[con2].embarked.value_counts()
```

```python
# isna only loads things with no value
df.head() #deck
```

```python
df.deck.value_counts() # From this sentence, We can not see Nan
```

```python
df.deck.value_counts(dropna = False)
```

```python
con3 =df.deck.isna()
df[con3]
```

```python
df[~con3] #~=not
```

#and(&), or(|)


*   Combine two or more conditions




```python
True and False
```

```python
True or False
```

```python
True & False
```

```python
True | False
```

```python
df=pd.DataFrame({'num':[1,2,3,4,5,6,7,8,9]})
```

```python
df
```

```python
df['col_1']=df.num >= 6
df['col_2']=df.num >= 3
df['col_and']=df.col_1 & df.col_2
df['col_or']=df.col_1 | df.col_2
df
```

#Exercise

1.   titanic dataset filtering
*   Filter data for passengers whose class is First and age is 30 or higher
*   Filter data for passengers with survived = 1 or embark_town = 'Southampton'
*   Filter data for passengers with fare > 100 and sex female







```python
import seaborn as sns
df2=sns.load_dataset('titanic')
```

```python
df2.head()
```

```python
filter1= (df2['class'] == 'First') #Parenthesize each of the conditions
filter1
```

```python
filter1= (df2['class'] == 'First') & (df2.age >=30)
filter1
```

```python
df2[filter1].age.value_counts().sort_index()
```

```python
df2[filter1]['class'].value_counts().sort_index()
```

```python
#Exercise 2
filter2 = (df2['survived']==1) | (df2['embark_town']=='Southampton')
filter2
```

```python
df2[filter2]
```

```python
df3=df2[filter2]
df3[df3['survived'] != 1].survived.value_counts()
```

```python
df3[df3['survived'] != 1].embark_town.value_counts()
```

```python
filter3=(df2['fare']>=100) & (df2['sex']=='female')
filter3
```

```python
df4=df2[filter3]
df4.head()
```

```python
df4[filter3].fare.value_counts()
```

```python
df4[filter3]['sex'].value_counts()
```
