#Filtering


*   Filtering in pd.DataFrame is the process of leaving only rows that meet a condition




```python
import pandas as pd
data = {
    'Name':['A','B','C','D'],
    'Age':[25,30,35,40],
    'City':['New York','Los Angeles','New York', 'Chicago']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
con = df.Age>30
df['age_over_30'] = con
df
```

```python
df[con]
```

```python
df[df.Age>30]
```

#between(), isin(), isna()

```python
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
df.head() #20 years old or older 22 years old or younger # between(20,22)
```

```python
con = df.age.between(20,22) #method
```

```python
df[con].age.value_counts()
```

```python
df.embarked.value_counts()
```

```python
con2 = df.embarked.isin(['S','C'])
df[con2].embarked.value_counts()
```

```python
# isna only loads things with no value
df.head() #deck
```

```python
df.deck.value_counts() # From this sentence, We can not see Nan
```

```python
df.deck.value_counts(dropna = False)
```

```python
con3 =df.deck.isna()
df[con3]
```

```python
df[~con3] #~=not
```
