# Index

*   A unique identifier used to distinguish rows


    *   df.set_index('col1', inplace = True) #Specify index
    *   df.reset_index(inplace = True) # basic integer index









```python
#taxis
import seaborn as sns
df = sns.load_dataset('taxis')
```

```python
df.head(1)
```

```python
df.set_index('color', inplace=True)
```

```python
df
```

```python
df.reset_index(inplace = True)
```

```python
df
```

#Sort


*   Ordered by specific criteria
*   sort_value(): By a specific column
    *   df.sort_values('col1',ascending=False, inplace =True)
    *   ascending = False #Descending order
    *   ascending = True #Ascending order

*   sort_index() : by a index
    *   df.sort_index(inplace = True)  





```python
df.sort_values('distance',ascending=False, inplace= True)
```

```python
df.head(10)
```

```python
df.sort_index(inplace= True)
```

```python
df.head(10)
```

1.   Using data frames to perform operations







```python
import pandas as pd
data = {
    'ProductID':[101, 102, 103, 104],
    'SaleDate':['2021-07-01','2021-07-03','2021-07-02','2021-07-04'],
    'Quantity':[3,2,5,4]
}

df=pd.DataFrame(data)
```



*   Set ProductID as an index in the dataframe
*   Organize data in descending order by Quantity
*   After that, reset to the default index



```python
df.set_index('ProductID', inplace=True)
```

```python
df
```

```python
df.sort_values('Quantity',ascending=False, inplace= True)
```

```python
df
```

```python
df.reset_index(inplace=True)
```

```python
df
```
