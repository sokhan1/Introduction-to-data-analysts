# to_datatime


*   Convert time-related data to datetime objects in python
*   Can handle various types of date and time data



    *   pd.to_datetime(df['col'])





```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/LectureNote1/taxis.csv')
df.head()
```

```python
df.info()
```

```python
df['pickup']=pd.to_datetime(df.pickup)
```

#to_period



*   Convert date/time data to a given period ('M','W','Y', etc.)

    *   df['col'].dt.to_period('period').dt.strftime(%Y-%m-%d)




```python
df.head()
df.info()
```

```python
df.pickup.dt.to_period('W')
```

```python
df['dropoff']=pd.to_datetime(df.dropoff)
df.dropoff.dt.to_period('W')
```

```python
df.pickup.dt.to_period('W').dt.strftime('%Y-%m-%d')
```

```python
df['pickup_w']=df.pickup.dt.to_period('W').dt.strftime('%Y-%m-%d')
```

```python
df.head()
```

```python
df.info()
```
