#pd.concat()


*   The concat function combines data frames in a vertical direction



    *   result = pd.concat([df1, df2, df3]
                            , ignore_index=True
                                )




```python
import pandas as pd

df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age':[25, 30]})
df2 = pd.DataFrame({'Name': ['Chris', 'David'], 'Age':[22, 27]})
df3 = pd.DataFrame({'Name': ['Eve', 'Frank'], 'Age':[28, 33]})
```

```python
result = pd.concat([df1, df2, df3]
                   , ignore_index=True
                   )
result
```

#pd.merge()



    *   pd.merge(left_df, right_df
                 ,left_on='left_key', right_on='right_key'
                 , how='join_type'
                )






*   inner Merge: combine only on common keys
*   Outer Merge: includes all keys, and fills with NaN if not matched
*   left merge: combine based on key in left table, padding with NaN if not matching
*   right merge:combine based on key in right table, fill with NaN if not matching
