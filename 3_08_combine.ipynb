#pd.concat()


*   The concat function combines data frames in a vertical direction



    *   result = pd.concat([df1, df2, df3]
                            , ignore_index=True
                                )




```python
import pandas as pd

df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age':[25, 30]})
df2 = pd.DataFrame({'Name': ['Chris', 'David'], 'Age':[22, 27]})
df3 = pd.DataFrame({'Name': ['Eve', 'Frank'], 'Age':[28, 33]})
```

```python
result = pd.concat([df1, df2, df3]
                   , ignore_index=True
                   )
result
```

#pd.merge()



    *   pd.merge(left_df, right_df
                 ,left_on='left_key', right_on='right_key'
                 , how='join_type'
                )






*   inner Merge: combine only on common keys
*   Outer Merge: includes all keys, and fills with NaN if not matched
*   left merge: combine based on key in left table, padding with NaN if not matching
*   right merge:combine based on key in right table, fill with NaN if not matching



```python
import pandas as pd

left = pd.DataFrame({
    'key_l': ['K0', 'K1', 'K2'],
    'col_1': ['A1', 'A2', 'A3'],
    'col_2': ['B1', 'B2', 'B3']
})
right = pd.DataFrame({
    'key_r': ['K1', 'K2', 'K3'],
    'col_1': ['C0', 'C1', 'C2'],
    'col_2': ['D0', 'D1', 'D2']
})
```

```python
left
```

```python
right
```

```python
merged = pd.merge(left, right
                  ,left_on = 'key_l', right_on='key_r'
                  , how='inner'
                  )
merged
```

```python
merged = pd.merge(left, right
                  ,left_on = 'key_l', right_on='key_r'
                  , how='outer'
                  )
merged
```

```python
merged = pd.merge(left, right
                  ,left_on = 'key_l', right_on='key_r'
                  , how='left'
                  )
merged
```

```python
merged = pd.merge(left, right
                  ,left_on = 'key_l', right_on='key_r'
                  , how='right'
                  )
merged
```

```python
merged = pd.merge(left, right
                  ,left_on = 'key_l', right_on='key_r'
                  , how='outer'
                  )
merged
merged['new_key'] = merged['key_r'].combine_first(merged['key_l'])
merged
```

#Cross Join



*   Generate all combinations of df1 and df2



```python
merged = pd.merge(left, right
                  #,left_on = 'key_l', right_on='key_r'
                  , how='cross'
                  )
merged
```

```python
import pandas as pd

employees = pd.DataFrame({
    'employee_id': ['E01', 'E02', 'E03', 'E04'],
    'name':['Alice', 'Bob', 'Charlie', 'David'],
    'department': ['HR', 'Engineering', 'IT', 'Marketing']

})

salaries = pd.DataFrame({
    'id':['E01','E02','E03'],
    'salary':[70000, 85000, 95000]
})
```



1.   employees의 'employee_id'열과 salaries의 'id'열을 기준으로 두 데이터 프레임을 결합하십시오. 이때, 'employee_id'와 'id'가 일치하는 행만 포함 되도록 내부결합(inner join)을 사용 하십시오.
2.   결합된 결과에서 'employee_id','name','department','salary'열만 포함 되도록 하십시오.


```python
employees
```

```python
salaries
```

```python
merge_ex = pd.merge(employees, salaries
                  ,left_on = 'employee_id', right_on='id'
                  , how='inner'
                  )
```
