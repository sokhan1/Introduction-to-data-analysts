#df.groupby()


*   Aggregation is the process of grouping data according to certain criteria and then aggregating it based on those criteria. You need to set the criteria for grouping and the criteria for the specific operation or calculation that will be used for aggregation.
*   process

    *   Splitting(분할)
    *   Applying(적용)
    *   Combining(결합)



    *   df.groupby( by= 'col_group')['col_agg'].agg_func().reset_index()








```python
import pandas as pd
df=pd.DataFrame({
    'group': ['A', 'A', 'B', 'B', 'C', 'C'],
    'value': [10, 10, 30, 40, 60, 60]
})
```

```python
df
```

```python
print('A:', df[df.group == 'A'])
print('B:', df[df.group == 'B'])
print('C:', df[df.group == 'C'])
```

#Applying(적용)

*   Apply the function to each group independently.
*   In Pandas, aggregation functions like sum, mean, max, and min are applied to numeric data by default (sum, max, etc. are applied but not used), whereas count and nunique functions are applied to all data types (numbers, strings, etc.).





```python
print('A:',df[df.group == 'A']['value'].sum())
print('B:',df[df.group == 'B']['value'].sum())
print('C:',df[df.group == 'C']['value'].sum())
```

# Combining(결합)

*   Combine results into a single data structure




```python
df.groupby( by= 'group')['value'].sum().reset_index()
```

#reset_index()


*   Make a index column



```python
df.groupby( by= 'group')['value'].sum()
```

##Aggregation(집계)


*   1. sum, mean, max, min: these functions work on numeric data (strings are an error)




```python
df.groupby( by= 'group')['value'].sum().reset_index()
```

```python
df.groupby( by= 'group')['value'].std().reset_index()
```

```python
df
```

```python
df.groupby( by= 'group')['value'].count().reset_index()
```

```python
import pandas as pd
df2=pd.DataFrame({
    'group': ['A', 'A', 'B', 'B', 'C', 'C'],
    'value': [10, 10, 30, 40, 60, 60],
    'value2': [None,'d','d','s',None,None]
})
```

```python
df2.groupby( by= 'group')['value2'].count().reset_index() #None is not counted
```

```python
df2.groupby( by= 'group')['value2'].nunique().reset_index() #Duplicate values do not exist
```

```python
import pandas as pd
df3=pd.DataFrame({
    'group': ['A', 'A', 'B', 'B', 'C', 'C'],
    'value': [10, 10, 30, 40, 60, 60],
    'value2': [None,'d','d','d',None,None]
})
```

```python
df3.groupby( by= 'group')['value2'].nunique().reset_index() #Duplicate values  exist
```

#groupby.agg()


*   Performing multiple aggregation operations


    *   df.groupby(by='group').agg(
        col_name1 = ('col1','agg_func')
        , col_name2 = ('col2','agg_func')
        ).reset_index()
        # agg_func -> sum,mean,max,min, ...






```python
df3.groupby(by='group')['value2'].nunique().reset_index()
```

```python
df3.groupby(by='group').agg(
    col_name1 = ('value2','nunique')
    , val1_sum = ('value','sum')
    ).reset_index()
```
