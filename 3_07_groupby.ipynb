#df.groupby()


*   Aggregation is the process of grouping data according to certain criteria and then aggregating it based on those criteria. You need to set the criteria for grouping and the criteria for the specific operation or calculation that will be used for aggregation.
*   process

    *   Splitting(분할)
    *   Applying(적용)
    *   Combining(결합)



    *   df.groupby( by= 'col_group')['col_agg'].agg_func().reset_index()








```python
import pandas as pd
df=pd.DataFrame({
    'group': ['A', 'A', 'B', 'B', 'C', 'C'],
    'value': [10, 10, 30, 40, 60, 60]
})
```

```python
df
```

```python
print('A:', df[df.group == 'A'])
print('B:', df[df.group == 'B'])
print('C:', df[df.group == 'C'])
```
