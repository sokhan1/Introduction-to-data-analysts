# Data Type


1.   Categorical Data: Data that can be divided into a limited number of categories, such as gender, blood type, and grade level.

2.   Numeric Data: Numerically represented and computable data

    *   Binary Data: Data consisting of only 0 and 1, such as whether you passed or failed a test, or whether you purchased an item. This is sometimes categorized as categorical data.
    *   Ratio Data: Data that allows for ratio comparisons between values, examples include height, weight, and price.
