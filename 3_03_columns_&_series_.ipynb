

2.   Dataset description
*   The dataset has the following columns


    *   CustomerID
    *   OrderDate
    *   Product:name
    *   Quantity
    *   Price


*   exercise


    *   change CustomerID to ID
    *   Delete OrderDate column
    *   Add TotalAmount column which is Quantity times Price   













```python
import pandas as pd

data={
    'CustomerID':[101, 102, 103, 104],
    'OrderDate':['2021-01-01','2021-01-02','2021-01-03','2021-01-04'],
    'Product':['Apple','Banana','Cherry','Date'],
    'Quantity':[5,3,4,2],
    'Price':[100,200,300,400]
}

df=pd.DataFrame(data)
```

```python
df.head()
```

```python
df.rename(columns = {'CustomerID':'ID' },inplace = True)
```

```python
df.drop(labels='OrderDate',axis = 1, inplace=True)
```

```python
df['TotalAmount'] = df.Quantity * df.Price
```
